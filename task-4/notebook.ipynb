```python
import pandas as pd
```

```python
# Demonstrating injestion of different formats
df_csv = pd.read_csv('addiction_population_data.csv')
df_json = pd.read_json('addiction_population_data.json')
df_txt = pd.read_csv('addiction_population_data.txt', sep='\t')

# Main dataframe to work with
df = df_csv
```

```python
# Normalizing Columns Names
df.columns = df.columns.str.strip().str.lower()
```

```python
# Fill missing text data
cols_to_fix = ['education_level', 'social_support', 'therapy_history']
df[cols_to_fix] = df[cols_to_fix].fillna('Unknown')
```

```python
print(df.isnull().sum())
```

```python
# View first five rows
print(df.head())

# Statistical summary of all columns (age, annual_income_usd)
print(df.describe())

# Count values for a specific category (e.g., employment_status)
print(df['employment_status'].value_counts())
```

```python
df.to_csv('cleaned_addiction_data.csv', index=False)
print("New CSV file saved")
```
